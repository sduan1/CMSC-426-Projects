```python
import math
from random import sample
import cv2
import numpy as np
import matplotlib.pyplot as plt
def imregionalmax(image, kernel=None):
    """Find the regional max of the image. An approximation of MATLAB's
    imregionalmax function. Result only differs when surrounding pixels
    have the same value as the center.

    Parameters:
    - image: the input image
    - kernel: the size of the neiborhood region, default is 3x3, i.e.
              neighboring 8 pixels.
    Returns:
    - a bitmask image, where '1' indicates local maxima.
    Author:
    - Yu Fang
    References:
    - https://github.com/bhardwajvijay/Utils/blob/master/utils.cpp
    - https://stackoverflow.com/questions/5550290/find-local-maxima-in-grayscale-image-using-opencv
    """
    # dialate the image so that small values are replaced by local max
    local_max = cv2.dilate(image, kernel)
    # non-local max pixels (excluding pixel w/ constant 3x3 neighborhood)
    # will be replaced by local max, so the values will increase. remove them.
    # so the result is either local max or constant neighborhood.
    max_mask = image >= local_max
    # erode the image so that high values are replaced by local min
    local_min = cv2.erode(image, kernel)
    # only local min pixels and pixels w/ constant 3x3 neighborhood
    # will stay the same, otherwise pixels will be replaced by the local
    # min and become smaller. We only take non-local min, non-constant values.
    min_mask = image > local_min
    # boolean logic hack
    #   (local max || constant) && (!local min && !constant)
    # = local max && !local min && !constant
    # = local max && !constant
    return (max_mask & min_mask).astype(np.uint8)
def est_homography(src, dest):
    """ Compute the homography matrix from (x_src, y_src) to (x_dest, y_dest).
    Parameters:
    - src: (x,y) coordinates of N source pixels, where coordinates are row vectors,
           so the matrix has dimension Nx2 (N>=4).
    - dest: (x,y) coordinates of N destination pixels, where coordinates are row vectors,
            so the matrix has dimension Nx2 (N>=4).
    Returns:
    - the homography matrix such that H @ [x_src, y_src, 1].T = [x_dest, y_dest, 1].T
    Author:
    - Yu Fang
    """
    N = src.shape[0]
    if N != dest.shape[0]:
        raise ValueError("src and diff should have the same dimension")
    src_h = np.hstack((src, np.ones((N, 1))))
    A = np.array([np.block([[src_h[n], np.zeros(3), -dest[n, 0] * src_h[n]],
                            [np.zeros(3), src_h[n], -dest[n, 1] * src_h[n]]])
                  for n in range(N)]).reshape(2 * N, 9)
    [_, _, V] = np.linalg.svd(A)
    # take the right singular vector x corresponding to the least singular value
    # s.t. ||Ax - 0||^2 is minimized
    return V.T[:, 8].reshape(3, 3)
def apply_homography(H, src):
    """ Apply the homography H to src
    Parameters:
    - H: the 3x3 homography matrix
    - src: (x,y) coordinates of N source pixels, where coordinates are row vectors,
           so the matrix has dimension Nx2 (N>=4).
    Returns:
    - src: (x,y) coordinates of N destination pixels, where coordinates are row vectors,
           so the matrix has dimension Nx2 (N>=4).
    Author:
    - Yu Fang
    """
    src_h = np.hstack((src, np.ones((src.shape[0], 1))))
    dest =  src_h @ H.T
    return (dest / dest[:,[2]])[:,0:2]

def drawMatches(image1, kp1, image2, kp2, idx_pairs):
    """A wrapper around OpenCV's drawMatches.
    
    Parameters:
    - image1: the first image
    - kp1: *matrix indices* of the keypoints from image 1
           (Nx2 numpy array, where N is the number of keypoints)
    - image2: the second image
    - kp2: *matrix indices* of the keypoints from image 2 
           (Nx2 numpy array, where N is the number of keypoints)
    - idx_pairs: pairs of matching indices, e.g. if kp1[3] 
                 matches kp2[5], then idx_pairs=[[3,5],...]
                 (Kx2 numpy array, where K is the number of matches)
    Returns:
    - an image showing matching points
    Author:
    - Yu Fang
    """
    # note that the coordinates are reversed because the difference
    # between matrix indexing & coordinates.
    keypt1 = [cv2.KeyPoint(coord[1], coord[0], 40) for coord in kp1.tolist()]
    keypt2 = [cv2.KeyPoint(coord[1], coord[0], 40) for coord in kp2.tolist()]
    matches = [cv2.DMatch(pair[0], pair[1], 0) for pair in idx_pairs.tolist()]
    return cv2.drawMatches(image1, keypt1, image2, keypt2, matches, None, flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

def pad_with(vector, pad_width, iaxis, kwargs):
     pad_value = kwargs.get('padder', 0)
     vector[:pad_width[0]] = pad_value
     vector[-pad_width[1]:] = pad_value
```

## Introduction

In this project we will be stitching images togather to create a panorama. The idea is that we detect keypoints in the images and match them accordingly. The idea is pretty similar to SIFT expect we are using much easier descriptors. The whole process can be boiled down to 5 steps: ANMS, Feature Descriptor, Feature matching, RANSAC homography, Image warping. The steps will be demonstrated 1 by 1 by implementing the algrithm on set 1.

## ANMS

The idea behind ANMS is that after harrison's corner detection we some times get a large number of corners. It is a terrible idea to use all of them. It would be better if we can find the best N corners and use them as descriptors. The ANMS algrithm does this. The ANMS algrithm takes all corners' coordinates and sort the corners by the minum distance to other corners. In this way the best corner should have the largest minimum distance to others. This means that in this area, this corner is the alpha. The full implementation and demonstration is shown below

```python
img = cv2.imread('../images/set1/2.jpg')
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
corners = cv2.cornerHarris(img,5,3,0.04)
corners[corners<1e-5] = np.mean(corners)

def anms(c, n):
    local_max = imregionalmax(c)
    
    cord_raw = np.argwhere(local_max)
    subset = c[local_max>0]
    num_max = len(cord_raw)
    r = np.array([np.inf]*num_max)
    for i in range(num_max):
        sup = cord_raw[subset[i]<0.9*subset]
        if len(sup) != 0:
            sup_norm = np.min(np.linalg.norm(sup-cord_raw[i],axis=1))
            if r[i]>sup_norm:
                r[i] = sup_norm
    best = np.argsort(-r)
    best_cord = np.array([cord_raw[i] for i in best])[0:n,:]
    
    return best_cord
    
    
        
res1 = anms(corners, 500)   
Y = res1[:,0]
X = res1[:,1]
plt.imshow(img, cmap='gray')
plt.scatter(X,Y,s=1, facecolors='none', edgecolors='r')
plt.show()
```

## Feature descriptors

In the previouse step we have successfully located the best 500 corners as our key points. However, the raw coordinate is not enough when we are comparing these keypoints to other images. We need to find a way of encoding data in it so we can actually compare them with keypoints from other images. Here we are using a simple descriptor. What we will be doing is to select a  40X40 submatrix centered at each keypoint's coordinate. Then we subsample them to a 8X8 submatrix and reshape them to 64X1 or 1X64 matrix and we use this as descriptors. The fully implementation is shown below

```python
padded_img = np.pad(img, 40,pad_with)   
sub_imgs = []
for [x, y] in res1+40:
    sub_img = padded_img[x-20:x+20, y-20:y+20]
    sub_imgs.append(sub_img)

sub_imgs = np.array(sub_imgs)
gb_sub_imgs = [cv2.GaussianBlur(i, (5,5),0) for i in sub_imgs]

sub_sampled_des1 = [np.reshape(i[::5,::5],(64,)) for i in gb_sub_imgs]
sub_sampled_des1 = np.array([(i-np.mean(i))/np.std(i) for i in sub_sampled_des1])
```

```python
padded_img2 = np.pad(img2, 40,pad_with)   
sub_imgs = []
for [x, y] in res2+40:
    sub_img = padded_img2[x-20:x+20, y-20:y+20]
    sub_imgs.append(sub_img)

sub_imgs = np.array(sub_imgs)
gb_sub_imgs = [cv2.GaussianBlur(i, (5,5),0) for i in sub_imgs]

sub_sampled_des2 = [np.reshape(i[::5,::5],(64,)) for i in gb_sub_imgs]
sub_sampled_des2 = np.array([(i-np.mean(i))/np.std(i) for i in sub_sampled_des2])
```

## Feature Matching

After we have found the descriptors of key points from both images we can start matching them with the help of draw_matches implemented by Yu Fang. The way we find the matching pairs is we take the norm of difference between key points from each image and the pairs with minimum norm is the match. We store the matched key points' index into a matrix called pairs. Then we take the ratio of best match (lowest distance) to the second best match (second lowest distance) and if this is below some ratio keep the matched pair otherwise reject it. Finally we pass img1 res1 img2 res2 pairs into draw_matches to visualize it. The full implementation is shown below

```python
img1 = cv2.imread('../images/set1/2.jpg',)
img2 = cv2.imread('../images/set1/3.jpg',)
pairs = []
for i in range(len(res1)):
    diff = sub_sampled_des1[i]-sub_sampled_des2
    ssd = np.linalg.norm(diff,axis=1)
    ssd_idx = np.argsort(ssd)[:2]
    if ssd[ssd_idx[0]]/ssd[ssd_idx[1]] < 0.5:
        pairs.append([i,ssd_idx[0]])
    
    

pairs = np.array(pairs)
# img1 = cv2.imread('../images/set1/2.jpg')
# img2 = cv2.imread('../images/set1/3.jpg')
#print(pairs)
plt.imshow(drawMatches(img1, res1, img2, res2, pairs))
plt.show()
```

## RANSAC to estimate homography

After successfully mathing the keypoints, we now want to warp one of the image in a way that the coordinates of kp1 matches coordinates of kp2. However, there may be mis-matched pairs. In order to reduce the impact of mis-matched pairs, we will be using a RANSAC algrithm to estimate the best homography matrix. The idea is that we sample 4 pairs from all matched pairs and estimate the homography with the helper function implemented by Yu Fang and apply this homography onto the image we are warping. Then we calculate the difference between H@img1 and img2. We want the sampled pair that gives a homography that minimizes the diff and we use that sampled pairs to estimate the homography. Implementation shown below

```python
res1[:,[0,1]] = res1[:,[1,0]]
res2[:,[0,1]] = res2[:,[1,0]]


sampled_pairs = np.array(sample(list(pairs), 4))

best_H = np.zeros((3,3))
best_ssd = math.inf

img1_fp = res1[pairs[:,0]]
img2_fp = res2[pairs[:,1]]
ssd_r = math.inf
for i in range(10000):
    
    sampled_pairs = np.array(sample(list(pairs), 4))
    
    p_i = sampled_pairs[:,0]
    p_i_p = sampled_pairs[:,1]
    src = res1[p_i]
    dest = res2[p_i_p]
    H = est_homography(src, dest)
    
    Hp = apply_homography(H, img1_fp)

    ssd_r = sum(np.linalg.norm(img2_fp-Hp, axis=1))
    if ssd_r < best_ssd:
        best_ssd = ssd_r
        best_H = H
```

## Image warping

After obtaining the estimated homography matrix, the only thing left to do is to warp the image and stitch them togather. Here I am using np.maximum to stitch thoes 2 images as it gives a better result than averaging. 

```python
img1_fp = res1[pairs[:,0]]
img2_fp = res2[pairs[:,1]]

warped_img2 = cv2.warpPerspective(img1,best_H , (np.shape(img)[1]*2,np.shape(img)[0]*2))
warped_img2[0:np.shape(img1)[0],0:np.shape(img1)[1]] = img2
plt.imshow(warped_img2,cmap='gray')
set1_23 = warped_img2
```

## Generalized Function

```python
def stitch_img(img1, img2):
    """
    img1 is the image gets warped onto img2
    
    """
    img1_color = img1
    img2_color = img2
    img1_gray = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
    img2_gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
    
    corners1 = cv2.cornerHarris(img1_gray,5,3,0.04)
    corners1[corners1<1e-5] = np.mean(corners1)
    
    corners2 = cv2.cornerHarris(img2_gray,5,3,0.04)
    corners2[corners2<1e-5] = np.mean(corners2)
    
    
    res1 = anms(corners1, 500)
    res2 = anms(corners2, 500)
    
    # Descriptors:
    padded_img1 = np.pad(img1_gray, 40,pad_with)   
    sub_imgs = []
    for [x, y] in res1+40:
        sub_img = padded_img1[x-20:x+20, y-20:y+20]
        sub_imgs.append(sub_img)

    sub_imgs = np.array(sub_imgs)
    gb_sub_imgs = [cv2.GaussianBlur(i, (5,5),0) for i in sub_imgs]

    sub_sampled_des1 = [np.reshape(i[::5,::5],(64,)) for i in gb_sub_imgs]
    sub_sampled_des1 = np.array([(i-np.mean(i))/np.std(i) for i in sub_sampled_des1])
    
    padded_img2 = np.pad(img2_gray, 40,pad_with)   
    sub_imgs = []
    for [x, y] in res2+40:
        sub_img = padded_img2[x-20:x+20, y-20:y+20]
        sub_imgs.append(sub_img)

    sub_imgs = np.array(sub_imgs)
    gb_sub_imgs = [cv2.GaussianBlur(i, (5,5),0) for i in sub_imgs]

    sub_sampled_des2 = [np.reshape(i[::5,::5],(64,)) for i in gb_sub_imgs]
    sub_sampled_des2 = np.array([(i-np.mean(i))/np.std(i) for i in sub_sampled_des2])
    
    # Get matched pairs:
    pairs = []
    for i in range(len(res1)):
        diff = sub_sampled_des1[i]-sub_sampled_des2
        ssd = np.linalg.norm(diff,axis=1)
        ssd_idx = np.argsort(ssd)[:2]
        if ssd[ssd_idx[0]]/ssd[ssd_idx[1]] < 0.6:
            pairs.append([i,ssd_idx[0]])



    pairs = np.array(pairs)
    plt.imshow(drawMatches(img1, res1, img2, res2, pairs))
    
    
    # Ransac
    res1[:,[0,1]] = res1[:,[1,0]]
    res2[:,[0,1]] = res2[:,[1,0]]
    sampled_pairs = np.array(sample(list(pairs), 4))

    best_H = np.zeros((3,3))
    best_ssd = math.inf

    img1_fp = res1[pairs[:,0]]
    img2_fp = res2[pairs[:,1]]
    ssd_r = math.inf
    for i in range(10000):

        sampled_pairs = np.array(sample(list(pairs), 4))

        p_i = sampled_pairs[:,0]
        p_i_p = sampled_pairs[:,1]
        src = res1[p_i]
        dest = res2[p_i_p]
        H = est_homography(src, dest)

        Hp = apply_homography(H, img1_fp)

        ssd_r = sum(np.linalg.norm(img2_fp-Hp, axis=1))
        if ssd_r < best_ssd:
            best_ssd = ssd_r
            best_H = H
            
    warped_img2 = cv2.warpPerspective(img1,best_H , (np.shape(img1)[1]*2,np.shape(img2)[0]*2))
    warped_img2[0:np.shape(img2)[0],0:np.shape(img2)[1]] = np.maximum(img2,warped_img2[0:np.shape(img2)[0],0:np.shape(img2)[1]])
    return warped_img2
```
