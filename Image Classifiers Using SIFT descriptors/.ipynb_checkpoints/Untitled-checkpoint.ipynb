```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Scale space

```python
test_img = cv2.imread('airplanes/image_0001.jpg', cv2.IMREAD_GRAYSCALE)
(oct1_h, oct1_w) = np.shape(test_img)
octave1 = [cv2.GaussianBlur(test_img, (5,5), 2**(2**i)/4) for i in range(4)]
dof1 = [octave1[i+1] - octave1[i] for i in range(3)]

test_img2 = cv2.resize(test_img,((int(np.shape(test_img)[1]/2), int(np.shape(test_img)[0]/2))))
octave2 = [cv2.GaussianBlur(test_img2, (5,5), 2**(2**i)/4) for i in range(4)]
dof2 = [octave2[i+1] - octave2[i] for i in range(3)]

test_img3 = cv2.resize(test_img,((int(np.shape(test_img2)[1]/2), int(np.shape(test_img2)[0]/2))))
octave3 = [cv2.GaussianBlur(test_img3, (5,5), 2**(2**i)/4) for i in range(4)]
dof3 = [octave3[i+1] - octave3[i] for i in range(3)]

test_img4 = cv2.resize(test_img,((int(np.shape(test_img3)[1]/2), int(np.shape(test_img3)[0]/2))))
octave4 = [cv2.GaussianBlur(test_img4, (5,5), 2**(2**i)/4) for i in range(4)]
dof4 = [octave4[i+1] - octave4[i] for i in range(3)]

test_img5 = cv2.resize(test_img,((int(np.shape(test_img4)[1]/2), int(np.shape(test_img4)[0]/2))))
octave5 = [cv2.GaussianBlur(test_img5, (5,5), 2**(2**i)/4) for i in range(4)]
dof5 = [octave5[i+1] - octave5[i] for i in range(3)]
```

```python
oct1_aprox_kps = []
mask = np.array([[True, True, True],[True, False, True],[True, True, True]])
for m in range(1, oct1_h-1):
    for n in range(1, oct1_w-1):
        max_val1 = np.max(dof1[1][m-1:m+2,n-1:n+2][mask])
        max_val2 = np.max(dof1[0][m-1:m+2,n-1:n+2])
        max_val3 = np.max(dof1[2][m-1:m+2,n-1:n+2])
        
        min_val1 = np.min(dof1[1][m-1:m+2,n-1:n+2][mask])
        min_val2 = np.min(dof1[0][m-1:m+2,n-1:n+2])
        min_val3 = np.min(dof1[2][m-1:m+2,n-1:n+2])
        if dof1[1][m][n] > max(max_val1, max_val2, max_val3) or dof1[1][m][n] < min(min_val1, min_val2, min_val3):
            oct1_aprox_kps.append([m, n])
    

oct1_aprox_kps = np.array(oct1_aprox_kps)
```
